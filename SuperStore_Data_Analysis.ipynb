```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=True)
```

```python
df["Order_Year"] = df["Order Date"].dt.year
df["Order_Month"] = df["Order Date"].dt.month
df["Order_Month_Name"] = df["Order Date"].dt.month_name()
df["Order_Quarter"] = df["Order Date"].dt.to_period("Q")
```

```python
#Total Sales & Profit
df["Sales"].sum()
df["Profit"].sum()
```

```python
#Average Sales & Profit
df[["Sales", "Profit"]].mean()
```

```python
#Category-wise Analysis
df.groupby("Category")[["Sales", "Profit"]].sum()
df.groupby("Sub-Category")[["Sales", "Profit"]].sum().sort_values("Profit", ascending=False)
```

```python
#Shipping Mode Analysis
df.groupby("Ship Mode")[["Sales", "Profit"]].sum()
```

```python
#Year-wise Trend
df.groupby("Order_Year")[["Sales", "Profit"]].sum()
```

```python
#Export Cleaned Dataset
df.to_csv("Superstore_Cleaned.csv", index=False)
```
